```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd
from glob import glob
```

```python
pickled_files = glob("real_life_traces/results/*.pkl")
print(pickled_files)
```

```python
def get_df(filename):
    with open(filename, "rb") as f:
        df = pickle.load(f)
    return df
```

```python
thousand_genome = glob("real_life_traces/results/1000genome*.pkl")
for file in thousand_genome:
    df_results = get_df(file)
    plt.scatter(df_results["num_tasks"], df_results["GD_cost"] / df_results["RLP_cost"], label='GD / RLP', alpha=0.5)
    plt.scatter(df_results["num_tasks"], df_results["ETF-H_cost"] / df_results["RLP_cost"], label='ETF-H / RLP', alpha=0.5)
    plt.title(f"{file[25:-4]}")
    plt.xlabel("number of tasks in DAG")
    plt.ylabel("ratio")
    plt.legend()
    plt.savefig(f"/home/vivek/Documents/Caltech/real_life/{file[25:-4]}.png", format="png")
    plt.show()
   

```
