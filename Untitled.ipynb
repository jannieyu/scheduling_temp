```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

```

## Fork

```python
num_machines = 3
df = pd.read_csv("fork_3_05042022_021237.csv")
plt.scatter(df["num_tasks"], df["LR_cost"]  / df["RLP_cost"], label='LR / RLP', alpha=0.5)
plt.axhline(1,c="k")
plt.title("LR / RLP \nnum_machines = " + str(num_machines))
plt.xlabel("Number of tasks in DAG")
plt.ylabel("Ratio")
plt.legend()
plt.show()

num_machines = 10
df = pd.read_csv("fork_10_05042022_030255.csv")
plt.scatter(df["num_tasks"], df["LR_cost"]  / df["RLP_cost"], label='LR / RLP', alpha=0.5)
plt.axhline(1,c="k")
plt.title("LR / RLP \nnum_machines = " + str(num_machines))
plt.xlabel("Number of tasks in DAG")
plt.ylabel("Ratio")
plt.legend()
plt.show()

num_machines = 50
df = pd.read_csv("fork_50_05042022_035414.csv")
plt.scatter(df["num_tasks"], df["LR_cost"]  / df["RLP_cost"], label='LR / RLP', alpha=0.5)
plt.axhline(1,c="k")
plt.title("LR / RLP \nnum_machines = " + str(num_machines))
plt.xlabel("Number of tasks in DAG")
plt.ylabel("Ratio")
plt.legend()
plt.show()

num_machines = 100
df = pd.read_csv("fork_100_05042022_044516.csv")
plt.scatter(df["num_tasks"], df["LR_cost"]  / df["RLP_cost"], label='LR / RLP', alpha=0.5)
plt.axhline(1,c="k")
plt.title("LR / RLP \nnum_machines = " + str(num_machines))
plt.xlabel("Number of tasks in DAG")
plt.ylabel("Ratio")
plt.legend()
plt.show()
```
